Importing Dependencies

```python
import numpy as np
import pandas as pd
import difflib #if the user commits a spelling mistake, it'll make the prediction either ways
from sklearn.feature_extraction.text import TfidfVectorizer #in order to convert text in numerical dataset
from sklearn.metrics.pairwise import cosine_similarity #in order to find similarities in vectorized formats
```

Data Collection and Pre-Processing

```python
#loading the data from the csv file to a pandas dataframe
movies_data=pd.read_csv('/content/movies.csv')
```

```python
(movies_data.head())
```

```python
#number of rows and column in the dataframe
movies_data.shape
```

```python
# selecting the relevant features for recommendation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string
for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining  all the 5 selected features
combined_features=movies_data['genres']+" "+movies_data['keywords']+" "+movies_data['tagline']+" "+movies_data['cast']+" "+movies_data['director']
```

```python
combined_features.head()
```

```python
#converting the text to feature vectors

vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
 #getting the similarity scores using cosine similarity
 similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie from the user
movie_name=input('Enter the movie name: ')
```

```python
#creating a list will all the movie-names given in the dataset
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title
index_of_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_movie)
```

```python
#getting a list of similar movies

similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
movies_data.iloc[31]
```

```python
len(similarity_score)
```

```python
#sortinff the movies based on their similaruty score
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#printing the name of similar movies based on the index
print('Movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

Movie Recommendation System

```python
movie_name=input('Enter the movie name: ')
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

# Check if find_close_match is empty before accessing elements
if find_close_match:
  close_match=find_close_match[0]
  index_of_movie=movies_data[movies_data.title==close_match]['index'].values[0]
  similarity_score=list(enumerate(similarity[index_of_movie]))
  sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
  print('Movies suggested for you: \n')
  i=1
  for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if(i<30):
      print(i,'.',title_from_index)
      i+=1
else:
  print("No close matches found for the movie name. Please check the spelling or try a different movie.")

```
